## <font color='#056938'><b>Descripción del Problema</b></font>

El jefe de Talento Humano de una prestigiosa universidad cuenta con una base de datos en un archivo **CSV** llamado `DataBase.csv`. Este archivo contiene la información de todos los estudiantes matriculados en el último semestre. Sin embargo, los datos están desorganizados, lo que dificulta la extracción de indicadores clave para una presentación sobre los resultados de la matrícula.

Se encomienda a un grupo de ingenieros el diseño e implementación de un **programa en Python** que procese esta información de manera eficiente y genere reportes útiles.

## <font color='#056938'><b>Estructura del Programa</b></font>

El programa debe implementar un **menú interactivo** en consola con las siguientes opciones:

1. **Procesamiento de los Registros**:
   - Preguntar el **nombre del archivo** a cargar (ej.: `DataBase.csv`).
   - Preguntar la **ruta absoluta** donde se almacenarán los resultados procesados.
   - Organizar los registros en una estructura de carpetas:  
     **`SEDE / FACULTAD / PROGRAMA ACADÉMICO`**  
     (Crear subcarpetas jerárquicas y guardar archivos procesados en el nivel correspondiente, ej.: CSVs filtrados por programa).

2. **Datos Generales de Registro**:
   - Mostrar:  
     a. Número de **sedes**, número de **programas académicos** y **cantidad total de estudiantes**.  
     b. Cantidad de registros discriminados por **género** (Hombres / Mujeres / Otro, si aplica).

3. **Gráficos**:
   - Generar y mostrar (usando `matplotlib` o similar):  
     a. **Diagrama de barras**: Hombres vs. Mujeres, discriminado por **sede** (cantidad de estudiantes).  
     b. **Gráfico de torta**: Porcentaje de estudiantes por **región**.  
     c. **Gráfico de torta**: Porcentaje de estudiantes por **municipio** (donde viven).  
     d. **Gráfico general**: Inspirado en el poster de Gapminder (incluir un gráfico similar o enlace interactivo).  
        [Referencia: Gapminder World Poster 2015](https://www.gapminder.org/downloads/updated-gapminder-world-poster-2015/).

4. **Ingresar Nuevos Registros**:
   - Preguntar la **ruta absoluta** del archivo nuevo.
   - Preguntar el **nombre del archivo**.
   - **Cargar y appendear** los datos al archivo principal (`DataBase.csv`), manteniendo consistencia.

5. **Tiempo de Ejecución del Programa**:
   - Mostrar el tiempo total de ejecución en formato **`HH:MM:SS`** (desde el inicio hasta la selección de esta opción).

6. **Salir**:
   - Terminar el programa de forma limpia (cerrar archivos, liberar recursos).
